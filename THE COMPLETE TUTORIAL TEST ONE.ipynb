# Financial Statement Analisis Tutorial

## Objectives for this practice:

Scenario: You are a Data Scientist working for a consulting firm. One of your
colleagues from the Auditing department has asked you to help them assess the
financial statement of organisation X.
You have been supplied with two vectors of data: monthly revenue and monthly
expenses for the financial year in question. Your task is to calculate the following
financial metrics:

1. [>>>](#Profit-for-each-month)
profit for each month
1. [>>>](#Profit-after-tax-for-each-month---Tax-rate-is-30%)profit after tax for each month (the tax rate is 30%)
1. [>>>](#Profit-margin-for-each-month---Equals-to-profit-after-tax-divided-by-revenue)profit margin for each month - equals to profit after tax divided by revenue
1. [>>>](#Good-months---Where-the-profit-after-tax-was-greater-than-the-mean-for-the-year)good months - where the profit after  tax was greater than the mean for the year
1. [>>>](#Bad-months---Where-the-profit-after-tax-was-lower-than-the-mean-for-the-year)bad months - where the profit after tax was less than the mean for the year
1. [>>>](#The-best-month---Where-the-profit-after-tax-was-max-for-the-year)the best month - where the profit after tax was max for the year
1. [>>>](#The-worst-month---Where-the-profit-after-tax-was-min-for-the-year)the worst month - where the profit after tax was min for the year

All results need to be presented as vectors.
Results for dollar values need to be calculated with 0.01 precision, but need to be
presented in Units of $1,000 (i.e. 1k) with no decimal points.
Results for the profit margin ratio need to be presented in units of % with no
decimal points.
Note: Your colleague has warned you that it is okay for tax for any given month to be
negative (in accounting terms, negative tax translates into a deferred tax asset)

### Data

```python
revenue = [14574.49, 7606.46, 8611.41, 9175.41, 8058.65, 8105.44, 11496.28, 9766.09, 10305.32, 14379.96, 10713.97, 15433.50]
expenses = [12051.82, 5695.07, 12319.20, 12089.72, 8658.57, 840.20, 3285.73, 5821.12, 6976.93, 16618.61, 10054.37, 3803.96]
```

### Profit for each month
[Back to Title](#SECTION-2---Fundamentals-of-Python)

```python
profit = list([])
for i in range(0, len(revenue)):
    profit.append(revenue[i] - expenses[i])
for i in range(0, len(profit)):
    profit[i] = round(profit[i],2)

```

### Profit after tax for each month - Tax rate is 30%

[Back to Title](#SECTION-2---Fundamentals-of-Python)

```python
profit_after_tax = list([])
for i in range(0, len(profit)):
    profit_after_tax.append(profit[i] + profit[i] * 0.30)
for i in range(0, len(profit)):
    profit_after_tax[i] = round(profit_after_tax[i], 2)
```

### Profit margin for each month - Equals to profit after tax divided by revenue
[Back to Title](#SECTION-2---Fundamentals-of-Python)

```python
profit_margin = list([])
for i in range(0, len(profit)):
    profit_margin.append(profit_after_tax[i] / revenue[i])
for i in range(0, len(profit_margin)):
    profit_margin[i] = round(profit_margin[i] * 100, 1)
```

### Good months - Where the profit after tax was greater than the mean for the year
[Back to Title](#SECTION-2---Fundamentals-of-Python)

```python
mean_pat = sum(profit_after_tax)/len(profit_after_tax) # Mean of profit after tax
good_months = list([])
for i in range(0, len(profit_after_tax)):
    good_months.append(profit_after_tax[i] > mean_pat)

```

### Bad months - Where the profit after tax was lower than the mean for the year
[Back to Title](#SECTION-2---Fundamentals-of-Python)

```python
bad_months = list([])
for i in range(0, len(profit_after_tax)):
    bad_months.append(profit_after_tax[i] < mean_pat)
```

### The best month - Where the profit after tax was max for the year
[Back to Title](#SECTION-2---Fundamentals-of-Python)

```python
best_month = list([])
for i in range(0, len(profit_after_tax)):
    best_month.append(profit_after_tax[i] == max(profit_after_tax))
```

### The worst month - Where the profit after tax was min for the year
[Back to Title](#SECTION-2---Fundamentals-of-Python)

```python
worst_month = list([])
for i in range(0, len(profit_after_tax)):
    worst_month.append(profit_after_tax[i] == min(profit_after_tax))
```

### Convert all calculations to units of thousand of dolars

```python
revenue_1000 = [round(i/1000, 2) for i in revenue]
expenses_1000 = [round(i/1000, 2) for i in expenses]
profit_1000 = [round(i/1000, 2) for i in profit]
profit_after_tax_1000 = [round(i/1000, 2) for i in profit_after_tax]
```

### Print the results!
[Back to Title](#SECTION-2---Fundamentals-of-Python)

```python
print("Revenue")
print(revenue_1000)
print("\nExpenses")
print(expenses_1000)
print("Profit for each month")
print(profit_1000)
print("\nProfit after tax for each month - Tax rate is 30%")
print(profit_after_tax_1000)
print("\nProfit margin for each month - Equals to profit after tax divided by revenue")
print(profit_margin)
print("\nGood month - Where the profit after tax was greater than the mean for the year")
print(good_months)
print("\nThe best month - Where the profit after tax was lower than the mean for the year")
print(bad_months)
print("\nThe best month - Where the profit after tax was max for the year")
print(best_month)
print("\nThe worst month - Where the profit after tax was min for the year")
print(worst_month)
```
